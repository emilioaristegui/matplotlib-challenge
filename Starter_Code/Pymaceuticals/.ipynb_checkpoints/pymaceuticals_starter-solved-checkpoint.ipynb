## Analysis

### Summary Statistics 

- Summary statistics for tumor volume based on different drug regimens indicates that:
- The mean tumor volume ranged from 40.22 (Ramicane) to 55.24 (Ketapril), with Capomulin having a mean of 40.68.
- The median tumor volume varied from 40.67 (Ramicane) to 53.70 (Ketapril), with Capomulin having a median of 41.56. 
- The tumor volume variance, ranged from 23.49 (Ramicane) to 68.55 (Ketapril), with Capomulin having a variance of 24.95. 
- The standard deviation varied from 4.85 (Ramicane) to 8.28 (Ketapril), with Capomulin having a standard deviation of 4.99.
- The standard error, ranged from 0.32 (Ramicane) to 0.60 (Ketapril), with Capomulin having a standard error of 0.33.

- The bar charts indicated that Capomuline and Ramicane had the most # of Observed Mouse Timepoints, while Ceftamin and Propriva had the least.
- The pie charts showed that distribution consisted of 51% male mice and 49% female.

- Capomuline and Ramicane had smaller final tumor volumes (mm3) than Infubinol and Ceftamin.

- The Line Plot showed that with the Capomuline treatment, mouse I509 decreased its tumor volume (mm3) from 45mm3 in day 0 to 41mm3 in day 45.
- Capomuline proved to be a successful treatment.

### Correlation and Regression

- The Scatter Plot showed that Average Tumor Volume (mm3) and Weight (g) had what appears to be a correlation between 0.70-0.90 positive correlation which is considered a high positive correlation. 
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import warnings
warnings.filterwarnings('ignore')

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
# pd.read_csv() function from pandas to read the CSV files and store the data in two DataFrame objects.
# Takes the file path as an argument and returns a DataFrame object
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
# how="left" parameter specifies that it performs a left join, which means all the rows 
# from study_results DataFrame are included in the merged DataFrame, 
# matching rows from the mouse_metadata DataFrame are added based on the "Mouse ID" column.
mouse_df = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
mouse_df.head()
```

```python
# INSTRUCTIONS
# Checking the number of mice.

# CODE AND EXPLANATION
# (unique) is a pandas Series method that returns an array of unique elements in a column.
# (nunique) is a pandas DataFrame method that returns the number of unique elements in a column.
mouse_df["Mouse ID"].nunique()
```

```python
# INSTRUCTIONS
# Our data should be uniquely identified by Mouse ID and Timepoint

# CODE AND EXPLANATION
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# .duplicated(subset=['Mouse ID','Timepoint'], keep=False) is a pandas DataFrame method 
# identifies duplicate rows based on specific columns. 
# Checks for duplicates in the "Mouse ID" and "Timepoint" columns of the mouse_df DataFrame.
# 'subset' parameter specifies the columns to consider for identifying duplicates, 
# keep=False parameter ensures that all duplicate rows are marked as False
mouse_df.loc[mouse_df.duplicated(subset=['Mouse ID','Timepoint'], keep=False)]["Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouse_df.loc[mouse_df.duplicated(subset=['Mouse ID','Timepoint'], keep=False)]
```

```python
# INSTRUCTIONS
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

# CODE AND EXPLANATION
# [mouse_df["Mouse ID"]!="g989"] creates a boolean mask that checks if each value in the "Mouse ID" column 
# is not equal to "g989". 
# Creates boolean Series with True for rows where the mouse ID is not "g989" 
# and False for rows where the mouse ID is "g989".
# .loc[] is used to filter the mouse_df DataFrame based on the boolean mask and selects rows where mouse ID =! "g989".
clean_mouse_df = mouse_df.loc[mouse_df["Mouse ID"]!="g989"]
clean_mouse_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# INSTRUCTIONS
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM (Standard Error of the Mean) of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

# CODE AND EXPLANATION
# 1. .groupby() method to group the data in the clean_mouse_df DataFrame by the "Drug Regimen" column.
# 2. It then calculates the mean, median, variance, standard deviation, 
# and standard error of the mean for the "Tumor Volume (mm3)" column within each group.
# 3. Results stored in Series objects: 
# mean_tumor_volume, median_tumor_volume, variance_tumor_volume, std_dev_tumor_volume, and std_err_tumor_volume.
# 4. Assembles Series into a single summary DataFrame called summary_df, 
# where each column corresponds to a specific summary statistic.
mean_tumor_volume = clean_mouse_df.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
median_tumor_volume = clean_mouse_df.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
variance_tumor_volume = clean_mouse_df.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
std_dev_tumor_volume = clean_mouse_df.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
std_err_tumor_volume = clean_mouse_df.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]
summary_df = pd.DataFrame({
    "Mean Tumor Volume": mean_tumor_volume,
    "Median Tumor Volume": median_tumor_volume,
    "Tumor Volume Variance": variance_tumor_volume,
    "Tumor Volume Std. Dev.": std_dev_tumor_volume,
    "Tumor Volume Std. Err.": std_err_tumor_volume
})
summary_df
```

```python
# INSTRUCTIONS
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line

# CODE AND EXPLANATION
# 1. .groupby() method to group the data in the clean_mouse_df DataFrame by the "Drug Regimen" column.
# 2. .agg() method to specify the summary statistics to calculate for the "Tumor Volume (mm3)" column for each group.
# 3. Inside the agg() method, a dictionary is passed where the key is the column name to aggregate ("Tumor Volume (mm3)") 
# and the value is a list of the summary statistics to calculate for that column ('mean', 'median', 'var', 'std', 'sem').
# 4. .agg() method applies the specified aggregation functions to the grouped data
# then calculates the mean, median, variance, standard deviation, and standard error of the mean for 
# the "Tumor Volume (mm3)" column within each drug regimen.
# 5. Results returned as a single-line summary DataFrame called summary_df2, 
# where the drug regimens are the index and the summary statistics are presented as columns.
summary_df2 = clean_mouse_df.groupby('Drug Regimen').agg({'Tumor Volume (mm3)': ['mean','median','var','std','sem']})
summary_df2
```

## Bar and Pie Charts

```python
# INSTRUCTIONS
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
# Different ways to create the same chart

# CODE AND EXPLANATION
# 1. .value_counts() method calculates the count of each unique value in the "Drug Regimen" column 
# of the clean_mouse_df DataFrame 
# The result is stored in the drug_regimen_count Series, where the drug regimens are the index and the 
# count of observed mouse timepoints is the value.
drug_regimen_count = clean_mouse_df["Drug Regimen"].value_counts()

# 2. .plot() method to create a bar plot of the drug_regimen_count Series. 
#    kind='bar' argument specifies that a bar plot should be created.
# 3. .xlabel() parameter sets the label for the x-axis as "Drug Regimen".
# 4. .ylabel() parameter sets the label for the y-axis as "# of Observed Mouse Timepoints".
drug_regimen_count.plot(kind='bar',xlabel='Drug Regimen',ylabel='# of Observed Mouse Timepoints')

# 5. plt.show() function from Matplotlib is called to display the generated bar plot. 
# plt.show() also gets rid of text
plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# INSTRUCTIONS
# Calculate the final tumor volume of each mouse across four of the treatment regimens: 
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

# CODE AND EXPLANATION
# 1. groupby() to calculate the maximum timepoint for each unique 'Mouse ID' in the 'clean_mouse_df' DataFrame 
# The result is stored in'last_timepoints' variable. 
# Finds the last recorded timepoint for each mouse.
last_timepoints = clean_mouse_df.groupby('Mouse ID')['Timepoint'].max()

# 2. It merges the 'clean_mouse_df' DataFrame with the 'last_timepoints' DataFrame based on the common columns 
# 'Mouse ID' and 'Timepoint'. 
# pd.merge function with how='right' parameter, merges both DataFrames and keeps only the rows
# from the 'clean_mouse_df' DataFrame that have matching values in the last_timepoints' DataFrame.
# The result of this merge is stored in the 'final_tumor_volume' variable.
final_tumor_volume = pd.merge(clean_mouse_df, last_timepoints, on=['Mouse ID', 'Timepoint'],how='right')
final_tumor_volume
```

```python
# INSTRUCTIONS 
# 1. Put treatments into a list for the loop and later for plot labels
# 2. Create empty list to fill with tumor volume data (for plotting)
# 3. Calculate the IQR and quantitatively determine if there are any potential outliers 
# 4. Locate the rows which contain mice on each drug and get the tumor volumes
# 5. Add subset 
# 6. Determine outliers using upper and lower bounds


# CODE AND EXPLANATION
# 1. The code creates a list called 'treatments' that contains the names of the four different treatments.
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# 2. The code creates an empty list to fill with tumor volume data (for plotting)
tumor_vol_data = []

# 3. The loop statement iterates over each element in the 'treatments' list.
for treatment in treatments:
    # 4. Within the loop, it filters the 'final_tumor_volume' DataFrame based on the condition 
    # that selects the rows where the 'Drug Regimen' column matches the current treatment.
    # The tumor volumes for the selected treatment are extracted from the 'Tumor Volume (mm3)' column 
    # using .loc[] accessor. 
    tumor_volumes = final_tumor_volume.loc[final_tumor_volume['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    
    # 5. .append() for the extracted tumor volumes to be appended to the 'tumor_vol_data' list
    tumor_vol_data.append(tumor_volumes)
    
    # 6. .quantile() to calculate quartiles, interquartile range (IQR), lower bound, and upper bound
    quartile1 = tumor_volumes.quantile(0.25)
    quartile3 = tumor_volumes.quantile(0.75)
    iqr = quartile3 - quartile1
    lower_bound = quartile1 - (1.5 * iqr)
    upper_bound = quartile3 + (1.5 * iqr)
    
    # Potential outliers are identified by selecting values that are either less than the lower bound or greater 
    # than the upper bound
    potential_outliers = tumor_volumes.loc[(tumor_volumes < lower_bound) | (tumor_volumes > upper_bound)]
    
    # If there are no potential outliers (i.e., the 'potential_outliers' DataFrame is empty), 
    # it prints a message indicating that there are no outliers for the current treatment.
    # uses .empty:
    if potential_outliers.empty:
        print(f"{treatment}'s potential outliers: None")
    else:
        print(f"{treatment}'s potential outliers:\n{potential_outliers}")
```

```python
# INSTRUCTIONS
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

# CODE AND EXPLANATION
# Put treatments into a list
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create an empty list to fill with tumor volume data
tumor_vol_data = []

# Get the tumor volume data for each treatment
for treatment in treatments:
    tumor_volumes = final_tumor_volume.loc[final_tumor_volume['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    tumor_vol_data.append(tumor_volumes)

# Create a figure and axis for the box plot
fig, ax = plt.subplots()
flierprops = dict(marker='o', markerfacecolor='red', markersize=12)

# Plot the box plot
# flierprops=
ax.boxplot(tumor_vol_data, flierprops=flierprops)

# Set the x-axis labels
ax.set_xticklabels(treatments)

# Set the y-axis label
ax.set_ylabel('Tumor Volume (mm3)')

# Set the title of the plot
ax.set_title('Distribution of Tumor Volume by Treatment Group')

# Show the plot
plt.show()
```

## Line and Scatter Plots

```python
# INSTRUCTIONS
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

# CODE AND EXPLANATION
# Select a single mouse treated with Capomulin
mouse_id = 'l509'  # Replace 'your_mouse_id' with the actual ID of the mouse

# It filters the 'clean_mouse_df' DataFrame based on two conditions using the logical AND operator (&).
# The first condition is clean_mouse_df['Drug Regimen'] == 'Capomulin', which checks if the value in the 
#'Drug Regimen' column is equal to 'Capomulin'.
# The second condition is clean_mouse_df['Mouse ID'] == mouse_id, which checks if the value in the 'Mouse ID' 
# column is equal to the specified 'mouse_id'.
# .loc accessor is used to perform the filtering based on the conditions. 
# The resulting DataFrame contains only the rows where both conditions are satisfied.
# The filtered DataFrame is assigned to the variable 'capomulin_data', 
# which will hold the subset of data that matches the conditions.
capomulin_data = clean_mouse_df.loc[(clean_mouse_df['Drug Regimen'] == 'Capomulin') & (clean_mouse_df['Mouse ID'] == mouse_id)]

# Create the line plot
plt.plot(capomulin_data['Timepoint'], capomulin_data['Tumor Volume (mm3)'], marker='o')

# Set the x-axis label
plt.xlabel('Timepoint (days)')

# Set the y-axis label
plt.ylabel('Tumor Volume (mm3)')

# Set the title of the plot
plt.title(f'Capomulin treatment of mouse I509')

# Show the plot
plt.show()
```

```python
# INSTRUCTIONS
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# CODE AND EXPLANATION
# Filter the data for the Capomulin regimen
capomulin_data = clean_mouse_df[clean_mouse_df['Drug Regimen'] == 'Capomulin']

# Calculate the average observed tumor volume for each mouse
avg_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Get the corresponding weights for each mouse
mouse_weights = capomulin_data.groupby('Mouse ID')['Weight (g)'].mean()

# Create the scatter plot of mouse weight vs. average observed tumor volume
plt.scatter(mouse_weights, avg_tumor_volume)

# Set the x-axis label
plt.xlabel('Mouse Weight (g)')

# Set the y-axis label
plt.ylabel('Average Tumor Volume (mm3)')

# Set the title of the plot
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin)')

# Show the plot
plt.show()
```

## Correlation and Regression

```python
# INSTRUCTIONS
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

# import statements allow you to access and use statistical functions from scipy.stats and the LinearRegression
# class from sklearn.linear_model in your code.
import scipy.stats as stats
from sklearn.linear_model import LinearRegression

# linearregress function, pulls out slope and intercept

# It filters the 'clean_mouse_df' DataFrame to select only the rows where the 'Drug Regimen' is 'Capomulin'. 
# The resulting DataFrame is assigned to the variable 'capomulin_data'.
capomulin_data = clean_mouse_df[clean_mouse_df['Drug Regimen'] == 'Capomulin']

# It calculates the average observed tumor volume for each mouse by grouping the 'capomulin_data' DataFrame
# by 'Mouse ID' and computing the mean of the 'Tumor Volume (mm3)' column. 
# The result is stored in the 'avg_tumor_volume' variable.
avg_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# It calculates the corresponding weights for each mouse by grouping the 'capomulin_data' DataFrame by 
# 'Mouse ID' and computing the mean of the 'Weight (g)' column. 
# The result is stored in the 'mouse_weights' variable.
mouse_weights = capomulin_data.groupby('Mouse ID')['Weight (g)'].mean()

# It calculates the correlation coefficient and p-value between 'mouse_weights' and 'avg_tumor_volume' 
# pearsonr() function from the 'stats' module. 
# The correlation coefficient is assigned to 'correlation_coef', and the p-value is assigned to 'p_value'.
correlation_coef, p_value = stats.pearsonr(mouse_weights, avg_tumor_volume)

# It creates a linear regression model using the LinearRegression class. 
# The model is fitted using 'mouse_weights' as the input features and 'avg_tumor_volume' as the target variable.
regression_model = LinearRegression()
regression_model.fit(mouse_weights.values.reshape(-1, 1), avg_tumor_volume.values.reshape(-1, 1))

# It retrieves the slope of the linear regression line using 'regression_model.coef_[0][0]' and 
# the intercept using 'regression_model.intercept_[0]'.
slope = regression_model.coef_[0][0]
intercept = regression_model.intercept_[0]

# It creates a scatter plot of 'mouse_weights' (x-axis) vs. 'avg_tumor_volume' (y-axis) 
# using the scatter function from 'matplotlib.pyplot'.
plt.scatter(mouse_weights, avg_tumor_volume)

# plt.plot() the linear regression line on the scatter plot using the plot function and 'regression_model.predict' 
# to generate the y-values for the regression line.
plt.plot(mouse_weights, regression_model.predict(mouse_weights.values.reshape(-1, 1)), color='red', label='Linear Regression')

# Set the x-axis label from 'matplotlib.pyplot' functions 
plt.xlabel('Mouse Weight (g)')

# Set the y-axis label from 'matplotlib.pyplot' functions
plt.ylabel('Average Tumor Volume (mm3)')

# Set the title of the plot from 'matplotlib.pyplot' functions
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin)')

# It adds the correlation coefficient and the equation of the linear regression line as text annotations
# to the plot using the text function.
plt.text(20, 36, f"Correlation coefficient: {correlation_coef:.2f}", fontsize=12)

# Add the equation of the linear regression line to the plot
plt.text(20, 38, f"y = {slope:.2f}x + {intercept:.2f}", fontsize=12)

# It adds a legend to the plot using the legend function.
plt.legend()

# It displays the plot using the show function from 'matplotlib.pyplot'.
plt.show()
```
